```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
```

```python
dataset = pd.read_csv('IMDB-Movie-Data.csv')
dataset.head
#dataset.shapea
display(dataset)
```

```python
IMB_dados = dataset.drop(['Description'], axis = 1)
IMB_dados.head
display(IMB_dados) 
```

```python
IMB_dados_só_atores =IMB_dados['Actors'].str.split(",",expand = True)
IMB_dados_só_atores.columns = ['Ator 1', 'Ator 2', 'Ator 3', 'Ator 4']

display(IMB_dados_só_atores)
```

```python
IMB_dados_só_generos =IMB_dados['Genre'].str.split(",",expand = True)
IMB_dados_só_generos.fillna(value=np.nan, inplace=True)
IMB_dados_só_generos.columns = ['Genero 1', 'Genero 2', 'Genero 3']

display(IMB_dados_só_generos)
```

```python
IMB_dados_só_ano_votação = IMB_dados[['Year', 'Votes']]
display(IMB_dados_só_ano_votação)
```

# Diretores que mais apareceram em toda lista

```python
# Finding frequency of elements 
diretor_counts = IMB_dados['Director'].value_counts() 

# Printing the five most common elements 
print(diretor_counts.nlargest(5)) 
```

Fazer a seleção dos mais visiveis de todos os filmes, pode ser um pouco falho, já q o mais citado pode estar aparecendo apenas nos filmes menos conhecido, logo vamos aplicar a mesma logica, mas apenas para os 100 primeiros filmes

```python
TopDiretores100= IMB_dados['Director'].head(100).value_counts()
print(TopDiretores100.nlargest(5))
```

```python
print("Esses são os Diretores TOp5 todo dataframe:")
print(diretor_counts.nlargest(5)) 

print("Esses são os Diretores TOp5 dos 100 primeiros:")
print(TopDiretores100.nlargest(5))
```

# Atores que mais aparecem na lista






```python
IMB_dados["Actors"] = IMB_dados["Actors"].str.split(", ")
# Contando as frequências
all_names = [name for sublist in IMB_dados["Actors"] for name in sublist]
name_counts1 = pd.Series(all_names).value_counts()

# pegando os 5 nomes mais frequentes
top_5teste = name_counts1.head(5)

# imprimindo os resultados
print(top_5teste)
```
